```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('FF_data')
print("connected to sqlite3 verion:", sqlite3.version)
```

```python
df = pd.read_sql_query(
     """SELECT APINumber
              ,WellName
              ,StateName
              ,StateNumber
              ,CountyName
              ,CountyNumber
              ,JobEndDate
              ,OperatorName
              
       from RegistryUpload"""
     , conn)
```

```python
# Review top five rows
df.head()
```

```python
# Convert JobEndDate from string to date and replace erros with blanks in column endDate
df['endDate'] = pd.to_datetime(df['JobEndDate'],errors='coerce')
```

```python
# See the added column in the last three rows
df.tail(3)
```

```python
# See all rows withe error in converting JobEndDate to date format
df[df['endDate'].isnull()]
```

```python
# Wells which finished fracking on a specific date
df[df['endDate'] == '2020-01-23']
```

```python
# Wells reported with future completion dates
df[df['endDate'] > 'now']
```
